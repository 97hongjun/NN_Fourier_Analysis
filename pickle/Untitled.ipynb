```python
import numpy as np
import matplotlib.pyplot as plt
import pickle
```

```python
with open ('eig_vals_d10_t10000_c0.1.pkl', 'rb') as handle:
    t10000 = pickle.load(handle)
with open ('eig_vals_d10_t20000_c0.1.pkl', 'rb') as handle:
    t20000 = pickle.load(handle)
with open ('eig_vals_d10_t22000_c0.1.pkl', 'rb') as handle:
    t22000 = pickle.load(handle)
with open ('eig_vals_d10_t30000_c0.1.pkl', 'rb') as handle:
    t30000 = pickle.load(handle)
```

```python
d = 10
appx = []
for i in range(1, 30000):
    term = t30000[-2]/(float(i)**(1.0 + 1.0/float(d)))
    appx.append(term)
appx = np.array(appx)
```

```python
e10000 = np.abs(t10000[::-1][1:] - appx[:9999])
e20000 = np.abs(t20000[::-1][1:] - appx[:19999])
e22000 = np.abs(t22000[::-1][1:] - appx[:21999])
e30000 = np.abs(t30000[::-1][1:] - appx[:29999])
```

```python
plt.figure(figsize=(20,10))
plt.plot(np.log10(e10000))
plt.plot(np.log10(e20000))
plt.plot(np.log10(e22000))
plt.plot(np.log10(e30000))
```

```python
plt.figure(figsize=(20,10))
plt.plot(np.log10(t10000[::-1]), label='N=10000')
plt.plot(np.log10(t20000[::-1]), label='N=20000')
plt.plot(np.log10(t22000[::-1]), label='N=22000')
plt.plot(np.log10(t30000[::-1]), label='N=30000')
plt.xlabel('K')
plt.ylabel('Log_10 Eigenvalue')
plt.title('Log_10 Eigenvalue Spectra for Kernel Matrices')
plt.plot(np.log10(appx), label='hypothesis (N=infinity)')
plt.legend()
plt.savefig('eigval_spectra_d10.png')
```
