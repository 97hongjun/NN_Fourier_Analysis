```python
import numpy as np
import matplotlib.pyplot as plt
import pickle
```

```python
with open ('eig_vals_d4_t1000_c0.1.pkl', 'rb') as handle:
    t1000 = pickle.load(handle)
with open ('eig_vals_d4_t10000_c0.1.pkl', 'rb') as handle:
    t10000 = pickle.load(handle)
with open ('eig_vals_d4_t15000_c0.1.pkl', 'rb') as handle:
    t15000 = pickle.load(handle)
with open ('eig_vals_d4_t30000_c0.1.pkl', 'rb') as handle:
    t30000 = pickle.load(handle)
```

```python
d = 4
appx = []
for i in range(1, 30000):
    term = t30000[-2]/(float(i)**(1.0 + 1.0/float(d)))
    appx.append(term)
appx = np.array(appx)
```

```python
e1000 = np.abs(t1000[::-1][1:] - appx[:999])
e10000 = np.abs(t10000[::-1][1:] - appx[:9999])
e15000 = np.abs(t15000[::-1][1:] - appx[:14999])
e30000 = np.abs(t30000[::-1][1:] - appx[:29999])
```

```python
plt.figure(figsize=(20,10))
plt.plot(np.log10(e1000))
plt.plot(np.log10(e10000))
plt.plot(np.log10(e15000))
plt.plot(np.log10(e30000))
```

```python
plt.figure(figsize=(20,10))
plt.plot(np.log10(t1000[::-1]), label='N=1000')
plt.plot(np.log10(t10000[::-1]), label='N=10000')
plt.plot(np.log10(t15000[::-1]), label='N=15000')
plt.plot(np.log10(t30000[::-1]), label='N=30000')
plt.xlabel('K')
plt.ylabel('Log_10 Eigenvalue')
plt.title('Log_10 Eigenvalue Spectra for Kernel Matrices')
plt.plot(np.log10(appx), label='hypothesis (N=infinity)')
plt.legend()
plt.savefig('eigval_spectra_d4.png')
```
